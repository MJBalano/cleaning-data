```python
# Handle any missing data in your dataframe.
df_above_90_perc_null =df[df.columns[(df.isna().mean()*100)>=90]]
#print(df_above_90_perc_null.column)
print(df.shape)
df = df.drop(axis='columns', columns=df_above_90_perc_null)
print(df.shape)
```

```python
# Investigate 0% null columns:gives the number of unique values in each col
#from class
df_zero_null = df[df.columns[(df.isna().mean() * 100)==0]]
print(df_zero_null.info())

print(df.shape)
uni_columns = []
for col in df_zero_null:
  print(f"{col}: {len(df_zero_null[col].unique())}")
  if len(df_zero_null[col].unique()) == 1:
        uni_columns.append(col)
df = df.drop(axis=1, columns=uni_columns)
print(df.shape)
```
